**Setting up the Baseline model with Resnet50**

```python
class OLIVESDataset(Dataset):
    def __init__(self, hf_dataset):
        self.dataset = hf_dataset
        self.transform = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.5], std=[0.5])
        ])

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        sample = self.dataset[idx]
        image = self.transform(sample['Image'].convert("L"))
        label = torch.tensor([
            sample['B1'], sample['B2'], sample['B3'],
            sample['B4'], sample['B5'], sample['B6']
        ], dtype=torch.float32)
        return image, label
```

```python
class OLIVESModel(nn.Module):
    def __init__(self, num_outputs=6):
        super(OLIVESModel, self).__init__()
        self.backbone = models.resnet50(pretrained=True)

        # Change first conv layer to accept 1 channel instead of 3
        self.backbone.conv1 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)

        # Replace final FC layer to output 6 biomarker predictions
        self.backbone.fc = nn.Linear(self.backbone.fc.in_features, num_outputs)

    def forward(self, x):
        return torch.sigmoid(self.backbone(x))  # sigmoid for multi-label
```

**Training the baseline model**

```python
def has_valid_labels(example):
    return all(example[f'B{i}'] is not None for i in range(1, 7))

filtered_dataset = olives_train.filter(has_valid_labels)

train_dataset = OLIVESDataset(filtered_dataset)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

model = OLIVESModel()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)


num_epochs = 10
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        outputs = model(images)
        loss = criterion(outputs, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {running_loss / len(train_loader):.4f}")
```

**Decided to implement focal loss to deal with the imbalanced dataset**

```python
class FocalLoss(nn.Module):
    def __init__(self, alpha=1.0, gamma=2.0, reduction='mean'):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.reduction = reduction

    def forward(self, inputs, targets):
        eps = 1e-8
        BCE_loss = nn.functional.binary_cross_entropy(inputs, targets, reduction='none')
        pt = torch.exp(-BCE_loss)

        focal_loss = self.alpha * (1 - pt) ** self.gamma * BCE_loss

        if self.reduction == 'mean':
            return focal_loss.mean()
        elif self.reduction == 'sum':
            return focal_loss.sum()
        else:
            return focal_loss
```

**Training with focal loss instead of default BCELogitLoss**

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
criterion = FocalLoss(alpha=1.0, gamma=2.0).to(device)
train_dataset = OLIVESDataset(filtered_dataset)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
num_epochs = 5
model = OLIVESModel()
model = model.to(device)
optimizer = optim.Adam(model.parameters(), lr=1e-4)
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        outputs = model(images)
        loss = criterion(outputs, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {running_loss / len(train_loader):.4f}")
```

**Filtering the data and normalizing BCVA and CST data to improve training**

```python
def has_valid_labels_and_features(example):
    return all(
        example.get(f'B{i}') is not None for i in range(1, 7)
    ) and example.get('BCVA') is not None and example.get('CST') is not None
filtered_dataset = olives_train.filter(has_valid_labels_and_features)

df_train = filtered_dataset.to_pandas()
# Sum positive samples for each biomarker
label_counts = df_train[[f'B{i}' for i in range(1, 7)]].sum()

N = len(df_train)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
pos_weights = (N - label_counts) / label_counts
pos_weights = torch.tensor(pos_weights, dtype=torch.float32).to(device)
# calculating mean and std for BCVA and CST before training
def compute_tabular_stats(dataset):
    bcva_vals = []
    cst_vals = []
    for sample in dataset:
        if sample['BCVA'] is not None and sample['CST'] is not None:
            bcva_vals.append(sample['BCVA'])
            cst_vals.append(sample['CST'])

    bcva_vals = np.array(bcva_vals)
    cst_vals = np.array(cst_vals)

    bcva_mean, bcva_std = bcva_vals.mean(), bcva_vals.std()
    cst_mean, cst_std = cst_vals.mean(), cst_vals.std()

    return bcva_mean, bcva_std, cst_mean, cst_std

# Example usage
bcva_mean, bcva_std, cst_mean, cst_std = compute_tabular_stats(filtered_dataset)
```

**Implementing a deeper model for the MLP**

```python
from torchvision.models import resnet50, ResNet50_Weights
class OLIVESHybridModel(nn.Module):
    def __init__(self, num_outputs=6):
        super(OLIVESHybridModel, self).__init__()

        # Load pretrained ResNet101 with 1 input channel
        self.cnn = resnet50(weights=ResNet50_Weights.DEFAULT)
        self.cnn.conv1 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)  # for grayscale input
        self.cnn.fc = nn.Identity()  # remove classifier, we'll use our own

        # Fully connected layers (tabular + image)
        self.fc = nn.Sequential(
            nn.Linear(2048 + 2, 256),  # 2048 = ResNet101 output, 2 tabular features
            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Linear(128, num_outputs)
        )

    def forward(self, image, tabular):
        x_img = self.cnn(image)  # (B, 2048)
        x = torch.cat([x_img, tabular], dim=1)
        return self.fc(x)
```

**Final decision to create an encoder using Resnet101 with partially frozen parameters and deep MLP**

```python
from sklearn.metrics import f1_score
from torch.optim.lr_scheduler import ReduceLROnPlateau
from torch.utils.data import DataLoader
from torchvision import transforms
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ----- MODEL -----
class BiomarkerClassifier(nn.Module):
    def __init__(self, encoder, num_outputs=6):
        super().__init__()
        self.encoder = encoder
        for param in self.encoder.parameters():
            param.requires_grad = False  # freeze encoder

        for name, param in self.encoder.named_parameters():
          if "layer4" in name or "fc" in name:
              param.requires_grad = True

        self.classifier = nn.Sequential(
            nn.Linear(2048 + 2, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Linear(128, num_outputs)
        )

    def forward(self, img, tabular):
        x_img = self.encoder(img)
        x = torch.cat([x_img, tabular], dim=1)
        return self.classifier(x)
class OLIVESDataset(torch.utils.data.Dataset):
    def __init__(self, hf_dataset, bcva_mean, bcva_std, cst_mean, cst_std):
        self.dataset = hf_dataset
        self.bcva_mean = bcva_mean
        self.bcva_std = bcva_std
        self.cst_mean = cst_mean
        self.cst_std = cst_std

        self.transform = transforms.Compose([
              transforms.Resize((224, 224)),
              transforms.ToTensor(),
              transforms.Normalize(mean=[0.5], std=[0.5])
        ])

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        sample = self.dataset[idx]

        image = self.transform(sample['Image'].convert("L"))

        # Normalize BCVA and CST
        bcva = (sample['BCVA'] - self.bcva_mean) / self.bcva_std
        cst = (sample['CST'] - self.cst_mean) / self.cst_std
        tabular = torch.tensor([bcva, cst], dtype=torch.float32)

        label = torch.tensor([
            sample['B1'], sample['B2'], sample['B3'],
            sample['B4'], sample['B5'], sample['B6']
        ], dtype=torch.float32)

        return image, tabular, label

train_dataset = OLIVESDataset(train_dataset, bcva_mean, bcva_std, cst_mean, cst_std)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2)
# Assume encoder is ResNet101 (already pretrained with contrastive loss)
encoder = torch.hub.load('pytorch/vision:v0.10.0', 'resnet101', pretrained=True)
encoder.fc = nn.Identity()
encoder.conv1 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)

model = BiomarkerClassifier(encoder=encoder).to(device)

# ----- OPTIMIZER + LOSS -----
criterion = nn.BCEWithLogitsLoss(pos_weight=pos_weights.to(device))
optimizer = optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-4)
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=3, verbose=True)

# ----- TRAINING LOOP -----
num_epochs = 30


for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0

    for batch_idx, (images, tabular, labels) in enumerate(train_loader):
        images, tabular, labels = images.to(device), tabular.to(device), labels.to(device)

        outputs = model(images, tabular)
        loss = criterion(outputs, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

        if (batch_idx + 1) % 200 == 0:
            avg = running_loss / (batch_idx + 1)
            print(f"Epoch {epoch+1} Batch {batch_idx+1} Loss: {avg:.4f}")

    avg_loss = running_loss / len(train_loader)
    print(f" Epoch {epoch+1}/{num_epochs} Avg Loss: {avg_loss:.4f}")
    scheduler.step(avg_loss)
```

```python
import numpy as np
from sklearn.metrics import f1_score
from torch.utils.data import DataLoader
from torchvision import transforms
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
encoder = torch.hub.load('pytorch/vision:v0.10.0', 'resnet101', pretrained=True)
encoder.fc = nn.Identity()
encoder.conv1 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)

model = BiomarkerClassifier(encoder=encoder).to(device)
model.load_state_dict(torch.load("/content/best_model.pth"))
model.eval()
val_dataset = OLIVESDataset(val_dataset, bcva_mean, bcva_std, cst_mean, cst_std)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
y_true, y_pred = [], []

with torch.no_grad():
    for images, tabular, labels in val_loader:
        images, tabular = images.to(device), tabular.to(device)
        outputs = torch.sigmoid(model(images, tabular))
        preds = (outputs.cpu().numpy() > 0.5).astype(int)

        y_pred.extend(preds)
        y_true.extend(labels.numpy())

y_true = np.array(y_true)
y_pred = np.array(y_pred)

# Compute F1 scores
f1_micro = f1_score(y_true, y_pred, average='micro')
f1_macro = f1_score(y_true, y_pred, average='macro')

print(f"Final Validation F1 Micro: {f1_micro:.4f}")
print(f"Final Validation F1 Macro: {f1_macro:.4f}")
```
