**Checking if the dataset is balanced or imbalanced**

```python
import pandas as pd
def has_valid_labels(example):
    return all(example[f'B{i}'] is not None for i in range(1, 7))

filtered_dataset = olives_train.filter(has_valid_labels)

df_train= filtered_dataset.to_pandas()

# Sum positive samples for each biomarker
label_counts = df_train[[f'B{i}' for i in range(1, 7)]].sum()

# Optionally: normalize by number of samples to get ratio
label_ratios = label_counts / len(df_train)

print("Label Counts:\n", label_counts)
print("\n Label Ratios:\n", label_ratios)
```

```python
from sklearn.model_selection import GroupShuffleSplit
import pandas as pd
def compute_tabular_stats(dataset):
    bcva_vals = []
    cst_vals = []
    for sample in dataset:
        if sample['BCVA'] is not None and sample['CST'] is not None:
            bcva_vals.append(sample['BCVA'])
            cst_vals.append(sample['CST'])

    bcva_vals = np.array(bcva_vals)
    cst_vals = np.array(cst_vals)

    bcva_mean, bcva_std = bcva_vals.mean(), bcva_vals.std()
    cst_mean, cst_std = cst_vals.mean(), cst_vals.std()

    return bcva_mean, bcva_std, cst_mean, cst_std
def has_valid_labels_and_features(example):
    return all(
        example.get(f'B{i}') is not None for i in range(1, 7)
    ) and example.get('BCVA') is not None and example.get('CST') is not None
def prepare_olives_dataset(dataset):
    dataset = dataset.filter(has_valid_labels_and_features)
    df = dataset.to_pandas()

    N = len(df)
    label_counts = df[[f'B{i}' for i in range(1, 7)]].sum()
    pos_weights = (N - label_counts) / (label_counts + 1e-6)
    pos_weights = torch.tensor(pos_weights.values, dtype=torch.float32)

    bcva_mean, bcva_std, cst_mean, cst_std = compute_tabular_stats(dataset)

    return dataset, pos_weights, bcva_mean, bcva_std, cst_mean, cst_std
filtered_dataset, pos_weights, bcva_mean, bcva_std, cst_mean, cst_std = prepare_olives_dataset(olives_train)

# Convert to DataFrame
df = filtered_dataset.to_pandas()
split = GroupShuffleSplit(test_size=0.2, n_splits=1, random_state=42)
train_idx, val_idx = next(split.split(df, groups=df["Patient_ID"]))

train_dataset = filtered_dataset.select(train_idx)
val_dataset = filtered_dataset.select(val_idx)
```

**Final decision to create an encoder using Resnet101 with partially frozen parameters and deep MLP**

```python
from sklearn.metrics import f1_score
from torch.optim.lr_scheduler import ReduceLROnPlateau
from torch.utils.data import DataLoader
from torchvision import transforms
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ----- MODEL -----
class BiomarkerClassifier(nn.Module):
    def __init__(self, encoder, num_outputs=6):
        super().__init__()
        self.encoder = encoder
        for param in self.encoder.parameters():
            param.requires_grad = False  # freeze encoder

        for name, param in self.encoder.named_parameters():
          if "layer4" in name or "fc" in name:
              param.requires_grad = True

        self.classifier = nn.Sequential(
            nn.Linear(2048 + 2, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Linear(128, num_outputs)
        )

    def forward(self, img, tabular):
        x_img = self.encoder(img)
        x = torch.cat([x_img, tabular], dim=1)
        return self.classifier(x)
class OLIVESDataset(torch.utils.data.Dataset):
    def __init__(self, hf_dataset, bcva_mean, bcva_std, cst_mean, cst_std):
        self.dataset = hf_dataset
        self.bcva_mean = bcva_mean
        self.bcva_std = bcva_std
        self.cst_mean = cst_mean
        self.cst_std = cst_std

        self.transform = transforms.Compose([
              transforms.Resize((224, 224)),
              transforms.ToTensor(),
              transforms.Normalize(mean=[0.5], std=[0.5])
        ])

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        sample = self.dataset[idx]

        image = self.transform(sample['Image'].convert("L"))

        # Normalize BCVA and CST
        bcva = (sample['BCVA'] - self.bcva_mean) / self.bcva_std
        cst = (sample['CST'] - self.cst_mean) / self.cst_std
        tabular = torch.tensor([bcva, cst], dtype=torch.float32)

        label = torch.tensor([
            sample['B1'], sample['B2'], sample['B3'],
            sample['B4'], sample['B5'], sample['B6']
        ], dtype=torch.float32)

        return image, tabular, label

train_dataset = OLIVESDataset(filtered_dataset, bcva_mean, bcva_std, cst_mean, cst_std)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2)
# Assume encoder is ResNet101 (already pretrained with contrastive loss)
encoder = torch.hub.load('pytorch/vision:v0.10.0', 'resnet101', pretrained=True)
encoder.fc = nn.Identity()
encoder.conv1 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)

model = BiomarkerClassifier(encoder=encoder).to(device)

# ----- OPTIMIZER + LOSS -----
criterion = nn.BCEWithLogitsLoss(pos_weight=pos_weights.to(device))
optimizer = optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-4)
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=3, verbose=True)

# ----- TRAINING LOOP -----
num_epochs = 30


for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0

    for batch_idx, (images, tabular, labels) in enumerate(train_loader):
        images, tabular, labels = images.to(device), tabular.to(device), labels.to(device)

        outputs = model(images, tabular)
        loss = criterion(outputs, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

        if (batch_idx + 1) % 200 == 0:
            avg = running_loss / (batch_idx + 1)
            print(f"Epoch {epoch+1} Batch {batch_idx+1} Loss: {avg:.4f}")

    avg_loss = running_loss / len(train_loader)
    print(f" Epoch {epoch+1}/{num_epochs} Avg Loss: {avg_loss:.4f}")
    scheduler.step(avg_loss)
```
