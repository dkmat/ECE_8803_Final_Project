```python
import torch.nn as nn
import torchvision.models as models
import torch.optim as optim
import numpy as np
import pandas as pd
import torch
from torch.utils.data import DataLoader
import torch.optim as optim
from torchvision import transforms
from torch.utils.data import Dataset
from sklearn.model_selection import GroupShuffleSplit
from sklearn.metrics import f1_score, classification_report, precision_score, recall_score
```

```python
class OLIVESDataset(Dataset):
    def __init__(self, hf_dataset):
        self.dataset = hf_dataset
        self.transform = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.5], std=[0.5])
        ])

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        sample = self.dataset[idx]
        image = self.transform(sample['Image'].convert("L"))
        label = torch.tensor([
            sample['B1'], sample['B2'], sample['B3'],
            sample['B4'], sample['B5'], sample['B6']
        ], dtype=torch.float32)
        return image, label
```

```python
class OLIVESModel(nn.Module):
    def __init__(self, num_outputs=6):
        super(OLIVESModel, self).__init__()
        self.backbone = models.resnet50(pretrained=True)

        # Change first conv layer to accept 1 channel instead of 3
        self.backbone.conv1 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)

        # Replace final FC layer to output 6 biomarker predictions
        self.backbone.fc = nn.Linear(self.backbone.fc.in_features, num_outputs)

    def forward(self, x):
        return torch.sigmoid(self.backbone(x))  # sigmoid for multi-label
```

```python
def has_valid_labels(example):
    return all(example[f'B{i}'] is not None for i in range(1, 7))

filtered_dataset = olives_train.filter(has_valid_labels)
df = filtered_dataset.to_pandas()
split = GroupShuffleSplit(test_size=0.2, n_splits=1, random_state=42)
train_idx, val_idx = next(split.split(df, groups=df["Patient_ID"]))

val_dataset = filtered_dataset.select(val_idx)
```

```python
val_dataset = OLIVESDataset(val_dataset)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=2)

# === 3. Load your trained model ===
model = OLIVESModel()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.load_state_dict(torch.load("/content/baseline_model.pth", map_location=device))
model.to(device)
model.eval()

# === 4. Run inference ===
y_true, y_pred = [], []

with torch.no_grad():
    for images, labels in val_loader:
        images = images.to(device)

        outputs = model(images)
        preds = (outputs.cpu().numpy() > 0.5).astype(int)

        y_pred.extend(preds)
        y_true.extend(labels.numpy())

y_true = np.array(y_true)
y_pred = np.array(y_pred)

# === 5. Compute Metrics ===
f1_micro = f1_score(y_true, y_pred, average='micro')
f1_macro = f1_score(y_true, y_pred, average='macro')
precision_micro = precision_score(y_true, y_pred, average='micro')
precision_macro = precision_score(y_true, y_pred, average='macro')
recall_micro = recall_score(y_true, y_pred, average='micro')
recall_macro = recall_score(y_true, y_pred, average='macro')

report = classification_report(
    y_true, y_pred,
    target_names=[f'B{i}' for i in range(1, 7)],
    output_dict=True
)

# === 6. Display the results ===
print("\n Final Validation Results:")
print(f"F1 Micro: {f1_micro:.4f}")
print(f"F1 Macro: {f1_macro:.4f}")
print(f"Precision Micro: {precision_micro:.4f}")
print(f"Precision Macro: {precision_macro:.4f}")
print(f"Recall Micro: {recall_micro:.4f}")
print(f"Recall Macro: {recall_macro:.4f}")

# Display per-class metrics
print("\nDetailed classification report:")
print(pd.DataFrame(report).transpose())
```

```python
class OLIVESDataset(torch.utils.data.Dataset):
    def __init__(self, hf_dataset):
        self.dataset = hf_dataset
        self.transform = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.5], std=[0.5])
        ])

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        sample = self.dataset[idx]

        image = self.transform(sample['Image'].convert("L"))

        tabular = torch.tensor([sample['BCVA'], sample['CST']], dtype=torch.float32)

        label = torch.tensor([
            sample['B1'], sample['B2'], sample['B3'],
            sample['B4'], sample['B5'], sample['B6']
        ], dtype=torch.float32)

        return image, tabular, label
```

```python
class OLIVESHybridModel(nn.Module):
    def __init__(self, num_outputs=6):
        super(OLIVESHybridModel, self).__init__()

        self.cnn = models.resnet50(pretrained=True)
        self.cnn.conv1 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)
        self.cnn.fc = nn.Identity()

        self.fc = nn.Sequential(
            nn.Linear(2048 + 2, 128),
            nn.ReLU(),
            nn.Linear(128, num_outputs),
            nn.Sigmoid()
        )

    def forward(self, image, tabular):
        x_img = self.cnn(image)
        x = torch.cat([x_img, tabular], dim=1)
        return self.fc(x)
```

```python
def has_valid_labels_and_features(example):
    return all(
        example.get(f'B{i}') is not None for i in range(1, 7)
    ) and example.get('BCVA') is not None and example.get('CST') is not None
filtered_dataset = olives_train.filter(has_valid_labels_and_features)

split = GroupShuffleSplit(test_size=0.2, n_splits=1, random_state=42)
train_idx, val_idx = next(split.split(df, groups=df["Patient_ID"]))
df = filtered_dataset.to_pandas()
val_dataset = filtered_dataset.select(val_idx)
```

```python
val_dataset = OLIVESDataset(val_dataset)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=2)

# === 3. Load your trained model ===
model = OLIVESHybridModel()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.load_state_dict(torch.load("/content/hybrid_model.pth", map_location=device))
model.to(device)
model.eval()

# === 4. Run inference ===
y_true, y_pred = [], []

with torch.no_grad():
    for images, tabular, labels in val_loader:
        images = images.to(device)
        tabular = tabular.to(device)
        outputs = torch.sigmoid(model(images, tabular))
        preds = (outputs.cpu().numpy() > 0.5).astype(int)

        y_pred.extend(preds)
        y_true.extend(labels.numpy())

y_true = np.array(y_true)
y_pred = np.array(y_pred)

# === 5. Compute Metrics ===
f1_micro = f1_score(y_true, y_pred, average='micro')
f1_macro = f1_score(y_true, y_pred, average='macro')
precision_micro = precision_score(y_true, y_pred, average='micro')
precision_macro = precision_score(y_true, y_pred, average='macro')
recall_micro = recall_score(y_true, y_pred, average='micro')
recall_macro = recall_score(y_true, y_pred, average='macro')

report = classification_report(
    y_true, y_pred,
    target_names=[f'B{i}' for i in range(1, 7)],
    output_dict=True
)

# === 6. Display the results ===
print("\n Final Validation Results:")
print(f"F1 Micro: {f1_micro:.4f}")
print(f"F1 Macro: {f1_macro:.4f}")
print(f"Precision Micro: {precision_micro:.4f}")
print(f"Precision Macro: {precision_macro:.4f}")
print(f"Recall Micro: {recall_micro:.4f}")
print(f"Recall Macro: {recall_macro:.4f}")

# Display per-class metrics
print("\nDetailed classification report:")
print(pd.DataFrame(report).transpose())
```

```python
class BiomarkerClassifier(nn.Module):
    def __init__(self, encoder, num_outputs=6):
        super().__init__()
        self.encoder = encoder
        for param in self.encoder.parameters():
            param.requires_grad = False  # freeze encoder

        for name, param in self.encoder.named_parameters():
          if "layer4" in name or "fc" in name:
              param.requires_grad = True

        self.classifier = nn.Sequential(
            nn.Linear(2048 + 2, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Linear(128, num_outputs)
        )

    def forward(self, img, tabular):
        x_img = self.encoder(img)
        x = torch.cat([x_img, tabular], dim=1)
        return self.classifier(x)
class OLIVESDataset(torch.utils.data.Dataset):
    def __init__(self, hf_dataset, bcva_mean, bcva_std, cst_mean, cst_std):
        self.dataset = hf_dataset
        self.bcva_mean = bcva_mean
        self.bcva_std = bcva_std
        self.cst_mean = cst_mean
        self.cst_std = cst_std

        self.transform = transforms.Compose([
              transforms.Resize((224, 224)),
              transforms.ToTensor(),
              transforms.Normalize(mean=[0.5], std=[0.5])
        ])

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        sample = self.dataset[idx]

        image = self.transform(sample['Image'].convert("L"))

        # Normalize BCVA and CST
        bcva = (sample['BCVA'] - self.bcva_mean) / self.bcva_std
        cst = (sample['CST'] - self.cst_mean) / self.cst_std
        tabular = torch.tensor([bcva, cst], dtype=torch.float32)

        label = torch.tensor([
            sample['B1'], sample['B2'], sample['B3'],
            sample['B4'], sample['B5'], sample['B6']
        ], dtype=torch.float32)

        return image, tabular, label
```

```python
def compute_tabular_stats(dataset):
    bcva_vals = []
    cst_vals = []
    for sample in dataset:
        if sample['BCVA'] is not None and sample['CST'] is not None:
            bcva_vals.append(sample['BCVA'])
            cst_vals.append(sample['CST'])

    bcva_vals = np.array(bcva_vals)
    cst_vals = np.array(cst_vals)

    bcva_mean, bcva_std = bcva_vals.mean(), bcva_vals.std()
    cst_mean, cst_std = cst_vals.mean(), cst_vals.std()

    return bcva_mean, bcva_std, cst_mean, cst_std
def has_valid_labels_and_features(example):
    return all(
        example.get(f'B{i}') is not None for i in range(1, 7)
    ) and example.get('BCVA') is not None and example.get('CST') is not None
def prepare_olives_dataset(dataset):
    dataset = dataset.filter(has_valid_labels_and_features)
    df = dataset.to_pandas()

    N = len(df)
    label_counts = df[[f'B{i}' for i in range(1, 7)]].sum()
    pos_weights = (N - label_counts) / (label_counts + 1e-6)
    pos_weights = torch.tensor(pos_weights.values, dtype=torch.float32)

    bcva_mean, bcva_std, cst_mean, cst_std = compute_tabular_stats(dataset)

    return dataset, pos_weights, bcva_mean, bcva_std, cst_mean, cst_std
filtered_dataset, pos_weights, bcva_mean, bcva_std, cst_mean, cst_std = prepare_olives_dataset(olives_train)

# Convert to DataFrame
df = filtered_dataset.to_pandas()
split = GroupShuffleSplit(test_size=0.2, n_splits=1, random_state=42)
train_idx, val_idx = next(split.split(df, groups=df["Patient_ID"]))
val_dataset = filtered_dataset.select(val_idx)
```

```python
val_dataset = OLIVESDataset(val_dataset,bcva_mean, bcva_std, cst_mean, cst_std)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=2)

# === 3. Load your trained model ===
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
encoder = torch.hub.load('pytorch/vision:v0.10.0', 'resnet101', pretrained=True)
encoder.fc = nn.Identity()
encoder.conv1 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)

model = BiomarkerClassifier(encoder=encoder).to(device)
model.load_state_dict(torch.load("/content/best_model.pth", map_location=device))
model.to(device)
model.eval()

# === 4. Run inference ===
y_true, y_pred = [], []

with torch.no_grad():
    for images, tabular, labels in val_loader:
        images = images.to(device)
        tabular = tabular.to(device)
        outputs = torch.sigmoid(model(images, tabular))
        preds = (outputs.cpu().numpy() > 0.5).astype(int)

        y_pred.extend(preds)
        y_true.extend(labels.numpy())

y_true = np.array(y_true)
y_pred = np.array(y_pred)

# === 5. Compute Metrics ===
f1_micro = f1_score(y_true, y_pred, average='micro')
f1_macro = f1_score(y_true, y_pred, average='macro')
precision_micro = precision_score(y_true, y_pred, average='micro')
precision_macro = precision_score(y_true, y_pred, average='macro')
recall_micro = recall_score(y_true, y_pred, average='micro')
recall_macro = recall_score(y_true, y_pred, average='macro')

report = classification_report(
    y_true, y_pred,
    target_names=[f'B{i}' for i in range(1, 7)],
    output_dict=True
)

# === 6. Display the results ===
print("\n Final Validation Results:")
print(f"F1 Micro: {f1_micro:.4f}")
print(f"F1 Macro: {f1_macro:.4f}")
print(f"Precision Micro: {precision_micro:.4f}")
print(f"Precision Macro: {precision_macro:.4f}")
print(f"Recall Micro: {recall_micro:.4f}")
print(f"Recall Macro: {recall_macro:.4f}")

# Display per-class metrics
print("\nDetailed classification report:")
print(pd.DataFrame(report).transpose())
```
