```python
import torch.nn as nn
import torchvision.models as models
import torch.optim as optim
import numpy as np
import pandas as pd
import torch
from torch.utils.data import DataLoader
import torch.optim as optim
from torchvision import transforms
from torch.utils.data import Dataset
```

```python
class OLIVESDataset(Dataset):
    def __init__(self, hf_dataset):
        self.dataset = hf_dataset
        self.transform = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.5], std=[0.5])
        ])

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        sample = self.dataset[idx]
        image = self.transform(sample['Image'].convert("L"))
        label = torch.tensor([
            sample['B1'], sample['B2'], sample['B3'],
            sample['B4'], sample['B5'], sample['B6']
        ], dtype=torch.float32)
        return image, label
```

```python
class OLIVESModel(nn.Module):
    def __init__(self, num_outputs=6):
        super(OLIVESModel, self).__init__()
        self.backbone = models.resnet50(pretrained=True)

        # Change first conv layer to accept 1 channel instead of 3
        self.backbone.conv1 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)

        # Replace final FC layer to output 6 biomarker predictions
        self.backbone.fc = nn.Linear(self.backbone.fc.in_features, num_outputs)

    def forward(self, x):
        return torch.sigmoid(self.backbone(x))  # sigmoid for multi-label
```

```python
class FocalLoss(nn.Module):
    def __init__(self, alpha=1.0, gamma=2.0, reduction='mean'):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.reduction = reduction

    def forward(self, inputs, targets):
        eps = 1e-8
        BCE_loss = nn.functional.binary_cross_entropy(inputs, targets, reduction='none')
        pt = torch.exp(-BCE_loss)

        focal_loss = self.alpha * (1 - pt) ** self.gamma * BCE_loss

        if self.reduction == 'mean':
            return focal_loss.mean()
        elif self.reduction == 'sum':
            return focal_loss.sum()
        else:
            return focal_loss
```

```python
from sklearn.model_selection import GroupShuffleSplit
def has_valid_labels(example):
    return all(example[f'B{i}'] is not None for i in range(1, 7))

filtered_dataset = olives_train.filter(has_valid_labels)
df = filtered_dataset.to_pandas()
split = GroupShuffleSplit(test_size=0.2, n_splits=1, random_state=42)
train_idx, val_idx = next(split.split(df, groups=df["Patient_ID"]))

train_dataset = filtered_dataset.select(train_idx)
train_dataset = OLIVESDataset(train_dataset)
val_dataset = filtered_dataset.select(val_idx)
val_dataset = OLIVESDataset(val_dataset)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

model = OLIVESModel()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

criterion = FocalLoss(alpha=1.0, gamma=2.0).to(device)
optimizer = optim.Adam(model.parameters(), lr=1e-4)


num_epochs = 30
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        outputs = model(images)
        loss = criterion(outputs, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {running_loss / len(train_loader):.4f}")
```
